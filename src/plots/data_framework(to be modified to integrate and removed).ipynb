```python
#Import the dataset and pandas, numpy
from sklearn import datasets
import pandas as pd
import numpy as np
 
# Load the diabetes dataset and create a dataframe
data = datasets.load_diabetes()
diabetes = pd.DataFrame(data.data, columns=data.feature_names)

# Take a glance at the description of the dataset:
# print(data.DESCR)

# Add the target variable to the dataframe
diabetes['target'] = data.target
# diabetes['gender'] = 'Male' if diabetes['sex'] > 0 else 'Female'
diabetes['gender'] = ['Male' if s > 0 else 'Female' for s in diabetes['sex']]

# Print the first 5 rows of some variables
print(diabetes[['age', 'gender', 'bmi', 'target']].head())
```

```python
data = datasets.load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)


# print(iris.DESCR)
iris['species'] = data.target_names[data.target]
print(iris.head())

iris['sepal length'] = ['long' if l >= np.mean(iris['petal length (cm)']) else 'short' for l in iris['petal length (cm)']]
iris['petal length (compared to average)'] = iris['petal length (cm)'] >= np.mean(iris['petal length (cm)'])

print(len(iris))
print(data.target_names)

# print(iris[['sepal length', 'sepal width (cm)', 'petal length (compared to average)', 'petal width (cm)', 'species']].head(5))
```

```python
from matplotlib import pyplot as plt

plt.style.use('ggplot')

plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'], color='k', marker='o', linewidth=3, label='sepal width (cm)')
plt.plot(iris['sepal length (cm)'], iris['petal length (cm)'], color='b', label='petal length (cm)')


plt.title('Iris sepal width vs. petal width')
plt.xlabel('sepal width (cm)')
plt.ylabel('petal width (cm)')

plt.legend()

plt.show()
```
