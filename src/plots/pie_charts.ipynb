```python
from matplotlib import pyplot as plt
import numpy as np

import sys
sys.path.append('..')
from util import *
from data import read_games
```

```python
# Data
import os
data_directory = os.getcwd() + '/../../bin/'
games_data = read_games(data_directory)

def get_count_on_columns(data, group_column_name, count_column_name):
    return data.groupby([group_column_name])[count_column_name].count().reset_index().rename(columns={count_column_name : 'Count'})

games_count_by_genre_and_year = games_data.groupby(['Genre', 'Release Year'])['User Rating'].mean()
games_count_by_genre = get_count_on_columns(games_data, 'Genre', 'Game Name')

print(games_count_by_genre)
```

```python
plt.style.use('fivethirtyeight')

plt.title('Games by genres')
plt.pie(x = games_count_by_genre['Count'], labels = games_count_by_genre['Genre'])
plt.show()
```

```python
## Select top 5 items with 
## 1. a black wedge props
## 2. rotate 90 degree and a explode on the top 3 items to emphasis on Action game:
## 3. exact percentage on each pie portions

explode = [0.2, 0.15, 0.1, 0, 0]
plt.title('Games by genres')
plt.pie(x = games_count_by_genre['Count'].head(5), explode = explode, shadow = True, labels = games_count_by_genre['Genre'].head(5), wedgeprops={'edgecolor': 'black'}, startangle = 90, autopct='%1.1f%%')
plt.show()
```
