```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets
data = datasets.load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)
iris['species'] = data.target_names[data.target]

print(iris.head())
```

### Figure: Container that holds the plots, A figure can have multiple plots (plt.gcf() - get current figure)
### Axes: actual plots.(plt.gca() - get current axes)

### 是一种更加科学的方式，用新创建的fig和axes对象来画图，而不是用静态对象的plt
### 这样不仅更加直观，也避免了使用`plt.gcf()` 和 `plt.gca()`

```python
fig, ax = plt.subplots()

ax.scatter(iris['sepal length (cm)'], iris['petal length (cm)'])
ax.set_title('Iris sepal length on their petal length')

print(ax)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
print(ax) ## A 2*1 array of axes

fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(nrows = 2, ncols = 2)
print(ax1) ## A 2*2 array of axes
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, sharex = True)
## Since two graghs share the same x-axis, only the bottom one will be shown with its mark


ax1.scatter(iris['sepal length (cm)'], iris['petal length (cm)'])
ax1.set_title('Iris sepal length on their petal length')

ax2.scatter(iris['sepal length (cm)'], iris['sepal width (cm)'])
ax2.set_xlabel('Iris sepal length (cm)')    
```
