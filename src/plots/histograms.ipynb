```python
#Import the dataset and pandas, numpy
from sklearn import datasets
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
data = datasets.load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)


# print(iris.DESCR)
iris['species'] = data.target_names[data.target]

iris['sepal length'] = ['long' if l >= np.mean(iris['petal length (cm)']) else 'short' for l in iris['petal length (cm)']]
iris['petal length (compared to average)'] = iris['petal length (cm)'] >= np.mean(iris['petal length (cm)'])

print(iris.head())
```

```python
plt.hist(iris['sepal length (cm)'], bins = 5, edgecolor = 'black')
```

```python
bins = [4.0, 5.0, 6.0, 7.0, 8.0]
median_length = iris['sepal length (cm)'].median()

plt.hist(iris['sepal length (cm)'], bins = bins, edgecolor = 'black')
plt.axvline(median_length, color='#fc4f30')

plt.title('Distribution of sepal length of iris')
plt.xlabel('Sepal length(cm)')
plt.ylabel('Appearance')
```
