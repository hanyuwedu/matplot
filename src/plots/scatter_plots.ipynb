```python
#Import the dataset and pandas, numpy
from sklearn import datasets
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
data = datasets.load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)

# print(iris.DESCR)
iris['species'] = data.target_names[data.target]
iris['sepal length'] = ['long' if l >= np.mean(iris['petal length (cm)']) else 'short' for l in iris['petal length (cm)']]
iris['petal length (compared to average)'] = iris['petal length (cm)'] >= np.mean(iris['petal length (cm)'])

print(iris.head())
```

```python
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'])
```

```python
plt.style.use('seaborn-v0_8')

plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], s = 100, c = 'green', marker = 'X', alpha = 0.75)
```

```python
def species_to_color(spice):
    if spice == 'setosa':
        return 'red'
    elif spice == 'versicolor':
        return 'blue'
    elif spice == 'virginica':
        return 'green'
    else:
        return 'black'

spices_color = iris['species'].apply(species_to_color)

plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], s = 100, c = spices_color)
```

```python
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], edgecolor = 'black', c = iris['sepal width (cm)'], cmap='Greens')
cbar = plt.colorbar()
cbar.set_label('Spices')
```
